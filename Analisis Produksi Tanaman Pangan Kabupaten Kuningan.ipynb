## Data Peta Kuningan

```python
peta = gpd.read_file('/home/yuma/Documents/Shapefile/Batas Kecamatan Indonesia/[LapakGIS.com] Batas Kecamatan Indonesia Update 2022 [38 Provinsi]/[LapakGIS.com] Batas_Kecamatan_BIG_PPBW.shp')
```

## Ketinggian Kecamatan

```python
kecamatan_mdpl = pd.read_csv('/home/yuma/Documents/My Project/Produksi Tanaman Pangan (Kab. Kuningan)/Analisis/Ketinggian Kecamatan Kabupaten Kuningan - Sheet1 (1).csv')
kecamatan_mdpl 
```

# JOIN DATA

## Join data berdasarkan kecamatan

```python
peta_pro_kng = peta_kng.join(produksi_kng.set_index('Kecamatan'), on='WADMKC')
peta_pro_kng
```

```python
peta_pro_kng.plot(column='Jagung', legend=True)
```

## Join Data Pro Max dan Peta

```python
peta_max_kng = peta_kng.join(max_kat_kecamatan.set_index('Kecamatan'), on='WADMKC')
peta_max_kng
```

```python
peta_max_kng['Komoditas'] = peta_max_kng['Komoditas'].fillna('')
peta_max_kng['Komoditas']
```

```python
ax = peta_max_kng.plot(column='Komoditas', 
                       legend=True, 
                       figsize=(15,15), 
                       cmap='Set2', edgecolor='white') 
```

## Join Data Ketinggian

```python
peta_kecamatan_mdpl = peta_kng.join(kecamatan_mdpl.set_index('Kecamatan'), on = 'WADMKC')
peta_kecamatan_mdpl
```

# MANIPULASI DATA

```python
peta_pro_kng['Jumlah'] = (peta_pro_kng['Jagung'] + 
                        peta_pro_kng['Kacang Hijau'] + 
                        peta_pro_kng['Kacang Tanah'] +
                        peta_pro_kng['Padi Sawah'] +
                        peta_pro_kng['Ubi Jalar'] +
                        peta_pro_kng['Ubi Kayu']) 
peta_pro_kng
```

```python
peta_pro_kng['Biji dan Kacang'] = (peta_pro_kng['Jagung'] +
                                   peta_pro_kng['Padi Sawah'] +
                                   peta_pro_kng['Kacang Hijau'] +
                                   peta_pro_kng['Kacang Tanah'])
peta_pro_kng['Umbi'] = (peta_pro_kng['Ubi Jalar'] +
                                   peta_pro_kng['Ubi Kayu'])
```

# VISUALISASI

```python
import matplotlib.patheffects as PathEffects
```

```python
peta_pro_kng['centroid'] = peta_pro_kng['geometry'].centroid
```

```python
ax = peta_pro_kng.plot(column='Jumlah',
                       cmap='Greens' , 
                       edgecolor='black', 
                       legend=True, 
                       figsize=(15,15), 
                       legend_kwds={'shrink': 0.3})

for point in peta_pro_kng.iterrows():
    ax.text(point[1]['centroid'].x,
            point[1]['centroid'].y,
            point[1]['WADMKC'],
            horizontalalignment='center',
            fontsize=10,
            path_effects=[PathEffects.withStroke(linewidth=3, foreground="white")])
ax.axis('off')

plt.savefig('kng2.png', transparent=True, dpi=300)
```
